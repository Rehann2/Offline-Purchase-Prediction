Number of visits to website vs number of purchases

```python
visits_per_product = visits["product_id"].value_counts().sort_index()
purchase_per_product = purchases["product_id"].value_counts().sort_index()
assert len(visits_per_product) == len(purchase_per_product)
```

```python
all(purchase_per_product.index == visits_per_product.index)
```

```python
pd.DataFrame({
    "visits": visits_per_product, 
    "purchases": purchase_per_product
    }).corr()
```
